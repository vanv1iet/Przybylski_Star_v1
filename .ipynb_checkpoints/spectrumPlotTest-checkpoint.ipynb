```python
def plotWavelength(wavelength, spectralArr, specLambda, lambdaRange):
    plt.plot(wavelength, spectralArr)
    plt.grid(True)
    plt.xlim(specLambda-lambdaRange/2, specLambda+lambdaRange/2)
    plt.title(str(specLambda)+" ")
    plt.xlabel('Wavelength ()')
    plt.ylabel('Intensity')
    plt.vlines(specLambda, ymin = 0, ymax = 14000)
    plt.show()
```

```python
#skip 1-79
#skip 378-571

data = pd.read_csv("H_lines_vacuum.csv")
data
```

```python
def air(lambda_vac, n = 1.0002849):
    lambda_air = lambda_vac/n
    return lambda_air
```

```python
def doppler(lambda_0, v_r):
    lambda_1 = lambda_0*(1+v_r/(3E5))
    return lambda_1
```

```python
ritz_wl_vac = np.array(data["ritz_wl_vac(A)"])
print(ritz_wl_vac)
```

```python
ritz_wl_air = []
for vacWL in ritz_wl_vac:
    ritz_wl_air.append(air(vacWL, n))
print(ritz_wl_air)
```

```python
ritz_wl_air_dopplerCorrected = []
for wavlen in ritz_wl_air:
    ritz_wl_air_dopplerCorrected.append(doppler(wavlen, 12.4))
print(ritz_wl_air_dopplerCorrected)
```

```python
count = 0
for doppWL in ritz_wl_air_dopplerCorrected:
    if doppWL >= start_wl and doppWL <= end_wl:
            plotWavelength(wavelength, spectArrA, doppWL, 2)
            count += 1
count
```
